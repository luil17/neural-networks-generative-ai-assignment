```python
# importing libarires
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
# loading dataset
fashion_mnist = tf.keras.datasets.fashion_mnist.load_data()
```

```python
# Getting training and test data from fashion mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist
X_train, y_train = X_train_full[:-5000], y_train_full[:-5000]
X_valid, y_valid = X_train_full[-5000:], y_train_full[-5000:]
```

```python
# Data shape
X_train.shape
X_test.shape
X_valid.shape
```

```python
y_train[1]
```

```python
# Making number smaller to make it faster
X_train = X_train/255.0
X_test = X_test/255.0
X_valid = X_valid/255.0
```

```python
# Class names or labels
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
"Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
# setting to random seed
tf.random.set_seed(42)

model = tf.keras.Sequential([
    # input layer
    tf.keras.layers.Flatten(input_shape=[28, 28,1]),
    # first layer 
    tf.keras.layers.Dense(110, activation="relu"),
    # second layer
    tf.keras.layers.Dense(50, activation="relu"),
    # output layer
    tf.keras.layers.Dense(10, activation="softmax")
])
# gradient decent
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
model.summary()
```

```python
# showing hidden layers and weights
model.layers
hidden1 = model.layers[1]
hidden1.name
weights, biases = hidden1.get_weights()
```

```python
# fitting model
model.fit(X_train, y_train, epochs=6)
```

```python
# Evaluation
test_loss, test_accruacy = model.evaluate(X_test, y_test)
```

using 80 neurons in the first layer 40 in the second layer and using sgd 

```python
# predicting 
predict = model.predict(X_test)
max_val = np.argmax(predict[0])
class_names[max_val]
```

```python
# generating new data
train_datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,  
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

test_datagen = ImageDataGenerator(rescale=1./255)

# training data
train_generator = train_datagen.flow(X_train, y_train, batch_size=32)
test_generator = test_datagen.flow(X_test, y_test, batch_size=32)

# Using the same layers and acitvation functions from the model 
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
```
